```python
import pandas as pd

# load dataset
df = pd.read_csv("/content/drive/MyDrive/Donar/Data_ CERF Donor Contributions and Allocations - contributions.csv")

# show first rows
df.head()
```

```python
df.columns   #Check columns
```

```python
df.info()  # Check basic info
```

```python
df = df[['donor', 'year', 'donorcommitment']] #Remove unnecessary columns
```

```python
df = df.rename(columns={
    'donor': 'Country',
    'year': 'Year',
    'donorcommitment': 'Contribution'
})  #Rename columns
```

```python
df = df.dropna() #Remove missing values
```

```python
df['Year'] = df['Year'].astype(int)
df['Contribution'] = df['Contribution'].astype(float) #Fix data types
```

```python
df.head()
df.info() #Check result
```

```python
df.to_csv("cleaned_data.csv", index=False) #Save cleaned dataset
```
